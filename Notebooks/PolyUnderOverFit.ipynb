```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(seed=42)

x_train = np.random.rand(14)*10
print(x_train)
x_val = np.random.rand(14)*10
x_train.sort()
x_val.sort()

print(x_val)
a = 0.2
b = -1
c = 1
y_train = np.array([a*x**2 + b*x + c + 2*np.random.random() for x in x_train])
y_val = np.array([a*x**2 + b*x + c + 2*np.random.random() for x in x_val])

y_lin = np.polyfit(x_train,y_train,1)
y_2 = np.polyfit(x_train,y_train,2)
y_10 = np.polyfit(x_train,y_train,10)

train_err_lin = np.linalg.norm(y_train - np.polyval(y_lin, x_train))/14
train_err_2 = np.linalg.norm(y_train - np.polyval(y_2, x_train))/14
train_err_10 = np.linalg.norm(y_train - np.polyval(y_10, x_train))/14

val_err_lin = np.linalg.norm(y_val - np.polyval(y_lin, x_val))/14
val_err_2 = np.linalg.norm(y_val - np.polyval(y_2, x_val))/14
val_err_10 = np.linalg.norm(y_val - np.polyval(y_10, x_val))/14

plt.figure(figsize=(10,6))
plt.scatter(x_train,y_train, label='training points',c='r')
plt.scatter(x_val, y_val, label='validation points',c='r',marker='+')
plt.plot(np.linspace(0,10,1000), np.polyval(y_lin, np.linspace(0,10,1000)), label='Linear')
plt.plot(np.linspace(0,10,1000), np.polyval(y_2, np.linspace(0,10,1000)), label='Quadratic')
plt.plot(np.linspace(0,10,1000), np.polyval(y_10, np.linspace(0,10,1000)), label='Deg10 Polynomial')
plt.legend()
plt.show()

print('Linear Training and Validation Error is: ', train_err_lin, val_err_lin)
print('Quadratic Training and Validation Error is: ',train_err_2, val_err_2)
print('Deg10 Polynomial Training and Validation Error is: ', train_err_10, val_err_10)
```
