```
import pandas as pd

import pickle


import dask.dataframe as dd
df = dd.read_csv("../Dataset2/*").compute()


X = df.reset_index().iloc[:,:400]
Y = df.reset_index().iloc[:,400:]

```

```
import numpy as np
gridpoints = []
for z in np.linspace(-2,2,11):
  for y in np.linspace(-2,2,11):
    for x in np.linspace(-2,2,11):
      gridpoints.append((x,y,z))
```

```
def insideCuboid(point, y):

  p0 = y[0]
  p1 = y[1]
  p2 = y[2]

  l0 = y[3]
  l1 = y[4]
  l2 = y[5]

  '''
    bool0 = (point[0]>p0) & (point[0]<p0+l0)
    bool1 = (point[1]>p1) & (point[1]<p1+l1)
    bool2 = (point[2]>p2) & (point[2]<p2+l2)
  '''

  bool0 = (point[0]>-l0/2) & (point[0]<l0/2)
  bool1 = (point[1]>-l1/2) & (point[1]<l1/2)
  bool2 = (point[2]>-l2/2) & (point[2]<l2/2)


  return bool0 & bool1 & bool2
```

```
Y_dots = []

for j in range(len(X)):
  if j%100==0:
    print(j)
  Y_dots.append([insideCuboid(gridpoint, Y.iloc[j]) for gridpoint in gridpoints])
```

```
y_voxels = pd.DataFrame(Y_dots, columns = gridpoints)
```

```
pd.concat([X.reset_index(drop=True), y_voxels],axis=1, ignore_index=True).to_csv('../VoxelsDataset/cuboidVoxelsNoOrigin.csv', index=False)
```
